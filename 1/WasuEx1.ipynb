```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def Psix(x):
    return np.sqrt(8/(2)^3) * np.sin(np.pi*x/2)
def Psiz(z):
    return np.sqrt(8/(2)^3) * np.sin(np.pi*z*2/2)

t = np.arange(0.0, 2.0, 0.01)

plt.figure(1)
plt.subplot(211)
plt.plot(t, Psix(t), 'k')

plt.figure(2)
plt.subplot(212)
plt.plot(t, Psiz(t), 'k')

plt.show()
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def Psi1(x,y):
    return np.sqrt(8/(1*2)) * np.sin(np.pi*x*1/1) * np.sin(np.pi*y*1/2)

def Psi2(x,y):
    return np.sqrt(8/(1*2)) * np.sin(np.pi*x*2/1) * np.sin(np.pi*y*1/2)

def Psi3(x,y):
    return np.sqrt(8/(1*2)) * np.sin(np.pi*x*1/1) * np.sin(np.pi*y*2/2)


fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.linspace(0, 1, 100)
y = np.linspace(0, 2, 100)

ax.plot(x, Psi1(x,y), zs=0, zdir='z', label='zs=0, zdir=z')

ax.plot(x, Psi2(x,y), zs=0, zdir='z', label='zs=0, zdir=z')

ax.plot(x, Psi3(x,y), zs=0, zdir='z', label='zs=0, zdir=z')


plt.show()

```
